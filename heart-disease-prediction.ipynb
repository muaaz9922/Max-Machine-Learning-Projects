```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/kaggle/input/d/mirzahasnine/heart-disease-dataset/heart_disease.csv')
df.head()
```

```python
df.cigsPerDay=df.cigsPerDay.fillna(9.003088619624615)
df.isnull().sum()
```

```python
df.BPMeds = df.BPMeds.fillna(0.02962962962962963)
df.isnull().sum()
```

```python
df.totChol = df.totChol.fillna(236.72158548233045)
df.isnull().sum()
```

```python
df.BMI.mean()
```

```python
df.BMI = df.BMI.fillna(25.80200758473572)
df.isnull().sum()
```

```python
df.glucose = df.glucose.fillna(81.96675324675324)
df.isnull().sum()
```

```python
df.Gender.unique()
```

```python
df = df.join(pd.get_dummies(df.Gender).astype(int))
df.head()
```

```python
df = df.drop(['Gender'],axis=1)
df.head()
```

```python
df.education.unique()
```

```python
df.education = df.education.map({'uneducated':0,'primaryschool':1,'graduate':2,'postgraduate':3})
df.head()
```

```python
df.prevalentStroke = df.prevalentStroke.map({'yes':1,'no':0})
df.head()
```

```python
df.columns
```

```python
df['Heart_ stroke'] = df['Heart_ stroke'].map({'No':0,'yes':1})
df.head()
```

```python
df.info()
```

```python
corr = df.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr,annot=True,cbar=True,cmap='plasma')
plt.show()
```

```python
#NOW check the distributions

#check dist of all columns
for col in df.columns:
  plt.figure(figsize=(10,6))
  sns.distplot(df[col])
  plt.show()
```

```python
#We see that the target column is imbalanced
df['Heart_ stroke'].value_counts()
```

```python
x = df.drop(['Heart_ stroke'],axis=1)
y = df['Heart_ stroke']

#Now we will try to balance the data using oversampling
#the data in target column is imbalanced. So we will now oversample it
#Just apply oversampling in target column
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=2)
x,y = smote.fit_resample(x,y.ravel())
np.unique(y,return_counts=True)



```

```python
#Apply the train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

#FEATURE SCALING
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)

#model selections
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

#objects
lr = LogisticRegression()
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
xgb = XGBClassifier()
svc = SVC()
knn = KNeighborsClassifier()
nb = GaussianNB()
lgb = LGBMClassifier()
cat = CatBoostClassifier()

#Fittings
lr.fit(x_train_scaled,y_train)
```

```python
rf.fit(x_train_scaled,y_train)
gb.fit(x_train_scaled,y_train)
xgb.fit(x_train_scaled,y_train)
svc.fit(x_train_scaled,y_train)
knn.fit(x_train_scaled,y_train)
nb.fit(x_train_scaled,y_train)
lgb.fit(x_train_scaled,y_train)
cat.fit(x_train_scaled,y_train)
```

```python
#preds
lrpred = lr.predict(x_test_scaled)
rfpred = rf.predict(x_test_scaled)
gbpred = gb.predict(x_test_scaled)
xgbpred = xgb.predict(x_test_scaled)
svcpred = svc.predict(x_test_scaled)
knnpred = knn.predict(x_test_scaled)
nbpred = nb.predict(x_test_scaled)
lgbpred = lgb.predict(x_test_scaled)
catpred = cat.predict(x_test_scaled)

#Evaluations
from sklearn.metrics import accuracy_score
lracc = accuracy_score(y_test,lrpred)
rfacc = accuracy_score(y_test,rfpred)
gbacc = accuracy_score(y_test,gbpred)
xgbacc = accuracy_score(y_test,xgbpred)
svcacc = accuracy_score(y_test,svcpred)
knnacc = accuracy_score(y_test,knnpred)
nbacc = accuracy_score(y_test,nbpred)
lgbacc = accuracy_score(y_test,lgbpred)
catacc = accuracy_score(y_test,catpred)

print('LOGISTIC REG',lracc)
print('RANDOM FOREST',rfacc)
print('GB',gbacc)
print('XGB',xgbacc)
print('SVC',svcacc)
print('KNN',knnacc)
print('NB',nbacc)
print('LIGHT GBM',lgbacc)
print('CATO',catacc)

```
