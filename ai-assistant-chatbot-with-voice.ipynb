```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime
import wikipedia
import pyjokes

listner = sr.Recognizer()
engine = pyttsx3.init()
voices = engine.getProperty('voice')
engine.setProperty('voice',voices[1].id)


def talk(text):
    engine.say('What can I do for you today?')
    engine.say('What are you upto today?')
    engine.runAndWait()

def take_command():
    try:
        with sr.Microphone() as source:
            print('listening....')
            voice = listner.listen(source)

            command = listener.recognize_google(voice)
            command = command.lower()
            if 'alexa' in command:
                command = command.replace('alexa','')
                print(command)
    except:
        pass 
    return command

def run_alexa():
    command = take_command()
    print(command)
    if 'play' in command:
        song = command.replace('play','')
        talk('playing....' + song)
        pywhatkit.playonyt(song)
    elif 'time' in command:
        time = datetime.datetime.now().strftime('%I:%M %p')
        print(time)
        talk('Current time is :', + time)
    elif 'who is ' im command:
        person = command.replace('Who is','')
        info = wikipedia.summary(person,1)
        print(info)
        talk(info)
    elif 'date' in command:
        talk('sorry, I have a headache')
    elif 'are you single' in command:
        talk('No! I am in a relationship with WIFI. Stay away from me or he will disconnect you from the world')
    elif 'joke' in command:
        talk(pyjokes.get_joke())
    else:
        talk('Please speak again')
    
while True:        
    run_alexa()
    
```
