```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
df.head()
```

```python
df.columns
```

```python
df['tags'] = df['director'] + ' '+ df['cast'] + ' '+df['listed_in'] +' '+ df['description']
df.head()
```

```python
df = df[['show_id','type','title','tags']]
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.reset_index()
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5700,stop_words='english')

from sklearn.metrics.pairwise import cosine_similarity
vector = cv.fit_transform(df['tags'].values.astype('U')).toarray()
similarity = cosine_similarity(vector)
```

```python
df[df['title']=='Zodiac'].index[0]
```

```python
similar_movies = sorted(list(enumerate(similarity[5696])),reverse=True,key=lambda vector:vector[1])
```

```python
for i in similar_movies[0:6]:
    print(df.iloc[i[0]].title)
    print(i[1])
```

```python
def recommend_movie(movie_name):
    movie_index = df[df['title']==movie_name].index[0]
    similar_movies = sorted(list(enumerate(similarity[5696])),reverse=True,key=lambda vector:vector[1])
    for i in similar_movies[1:6]:
        print(df.iloc[i[0]].title+'with similarity of'+str(round(i[1]*100,2)))
        
    
recommend_movie('Ganglands')
```
